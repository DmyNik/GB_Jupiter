```python
# Условие 4: 4 задача
# Ответьте на несколько вопросов
# 4.1 Сколько в среднем стоят дома, у которых 2 спальни?
# 4.2 Какая в среднем общая площадь домов, у которых стоимость больше 600 000?
# 4.3 Как много домов коснулся ремонт?
# 4.4 Насколько в среднем стоимость домов с оценкой grade домов выше 10 отличается от стоимости домов с оценкой grade меньше 4?

# Сколько в среднем стоят дома, у которых 2 спальни:
df_new = df[df['bedrooms'] == 2]
print('Дома, у которых 2 спальни в среднем стоят: ' + str(round(df_new['price'].mean(), 2)) + '$')

# Какая в среднем общая площадь домов, у которых стоимость больше 600 000:
df_new = df[df['price'] > 600000]
print('В среднем общая площадь домов, у которых стоимость больше 600 000: ' + str(round(df_new['sqft_lot'].mean(), 2)) + '$')

# Как много домов коснулся ремонт?
df_new = df[df['yr_renovated'] > 0]
print('Домов, которых коснулся ремонт: ' + str(df_new['id'].count()))

# Насколько в среднем стоимость домов с оценкой grade домов выше 10 отличается от стоимости домов с оценкой grade меньше 4?
grade_10up = df[df['grade'] > 10]
grade_10up = grade_10up['price'].mean()
grade_4down = df[df['grade'] < 4]
grade_4down = grade_4down['price'].mean()
print('Средняя стоимость домов с оценкой grade домов выше 10 отличается от стоимости домов с оценкой grade меньше 4 на: ' + str(round(grade_10up - grade_4down)) + '$')
```
